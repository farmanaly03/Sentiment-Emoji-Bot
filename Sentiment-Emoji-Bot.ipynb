```python
from transformers import pipeline
```

```python
import random
```

```python
sentiment_pipeline=pipeline("sentiment-analysis")
```

```python
# Mapping based on detected sentiment
emoji_dict = {
    "POSITIVE": ["üòä", "üòÑ", "üòÅ", "ü•≥", "ü§©", "‚ù§", "üëç"],
    "NEGATIVE": ["üò¢", "üò†", "üòû", "üíî", "üò≠", "üëé"],
    "NEUTRAL": ["üòê", "ü§î", "üôÉ", "üò∂", "ü´•"]
}


# Optional: Add basic keyword-based emotion mapping
keyword_emoji = {
    "happy": ["üòÑ", "üòä", "ü•∞"],
    "love": ["‚ù§", "üòç", "üòò"],
    "angry": ["üò†", "üò°", "ü§¨"],
    "sad": ["üò¢", "üò≠", "üòû"],
    "fun": ["ü§£", "üòú", "üéâ"],
    "excited": ["ü§©", "ü•≥", "üòÉ"]
}
```

```python
def generate_emojis(text):
    result = sentiment_pipeline(text)[0]
    label = result['label'].lower()

    # Get emojis from sentiment
    emoji_list = emoji_dict.get(label, [""])
    sample_size = min(2, len(emoji_list))  # avoid ValueError
    sentiment_emojis = random.sample(emoji_list, sample_size)

    # Get keyword-based emojis (if you use this)
    keyword_emojis = []
    for keyword in keyword_emoji:
        if keyword in text.lower():
            keyword_emojis.extend(keyword_emoji[keyword])

    # Combine all
    return sentiment_emojis + keyword_emojis
```

```python
input_text = input("Enter your mood or sentence: ")
emojis = generate_emojis(input_text)
print(f"Your emoji mood: {emojis}")
```
